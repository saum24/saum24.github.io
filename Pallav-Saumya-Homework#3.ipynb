### Solution:

   1. Relationship between the year acquired and the year constructed

```python
# importing the libraries

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Reading the CSV file
# Also, setting the '0' value as NaN for 'Square Footage','Year Acquired' & 'Year Constructed' columns

df1=pd.read_csv('building_inventory.csv',
                na_values={'Square Footage': 0,
                           'Year Acquired': 0,
                           'Year Constructed': 0})

df1.describe() # row count=8862
```

```python
# no '0' values in the data frame for below columns

print(df1['Year Acquired'].min())
print(df1['Year Constructed'].min())
print(df1['Square Footage'].min())
```

```python
#retreiving the rows where below columns have 'NA' (null) values

bool_series1=pd.isnull(df1['Year Acquired'])
bool_series2=pd.isnull(df1['Year Constructed'])
bool_series3=pd.isnull(df1['Square Footage'])

df1[bool_series1] #rows where 'Year Acquired' has NA value 

```

```python
df1[bool_series2] #rows where 'Year Constructed' has NA value 
```

```python
df1[bool_series3] #rows where 'Square Footage' has NA value 
```

```python
df_yr=df1
len(df_yr) # total number of rows in the copy data frame
```

```python
# dropping the rows where 'Year Acquired' and 'Year Constructed' have null values & saving the data frame at the same time
df_yr.dropna(axis = 0,subset=['Year Acquired','Year Constructed'],inplace=True)

# Verifying 'NA' rows are dropped
len(df_yr) #updated row count 
```

```python
# plotting the first graph Year Acquired VS Year Constructed

# plotting a figure
plt.figure()

#scatter plot b/n Year Acquired & Year Constructed, point size=5, and plot label
plt.scatter(df_yr['Year Acquired'],df_yr['Year Constructed'], s=5,label='Acquired VS Constructed') 

# labelling axes
plt.xlabel("Year Acquired")
plt.ylabel("Year Constructed")

# including grid in plot
plt.grid()

# plotting the legend
plt.legend()

# fiving title to graph
plt.title('Graph 1')

# displyaing the graph
plt.show()
```

In most cases, the Acquired Year is same as Constructed Year. Thus, there are many points on the y=x line. However, there are certain cases where the years are not same.

###   2. Total square footage as a function of congressional district ("Congress Dist")
###   3.  Average square footage per floor as a function of congressional district

```python
df_sf=df1

# dropping the rows where 'Square Footage' have null values ('NA') & saving the data frame at the same time
df_sf.dropna(axis = 0,subset=['Square Footage'],inplace=True)
```

```python
# Creating a new data frame for plotting the other two graphs
# the data frame used did not have any NA values for Square Footage column

#Grouping by Congress Dist

temp1=df_sf.groupby("Congress Dist")["Square Footage"].sum()  # Total Square Footage calculated for each Congress Dist
temp2=df_sf.groupby("Congress Dist")["Square Footage"].mean() # Mean of Square Footage calculated for each Congress Dist
temp3=df_sf.groupby("Congress Dist")["Total Floors"].sum()    # Total Floors calculated for each Congress Dist
 
# data frame created    
df2=pd.DataFrame({
    'Congress Dist':temp1.index,
    'Total Sq Ft':temp1.values,
    'Mean Sq Ft':temp2.values,
    'Total Floors':temp3.values,
    'Avg Sq Ft/Floor':temp2.values/temp3.values
})

print(df2)  # the new dataframe



```



A data frame with columns **Congressional District**, **Total square footage**, **Average square footage**, **Total Floors** and **Average square footage per floor** created.



```python
# Plotting the other two graphs

# Plotting a figure
plt.figure(num=1,figsize=(20.0,15.0))

# plotting a subplot for Total square footage as a function of congressional district
plt.subplot(221)
plt.plot(df2['Congress Dist'],df2['Total Sq Ft'],'C2',lw=0.5,marker = '*',label='Total Sq Ft')
plt.xlabel("Congress Dist")
plt.ylabel("Total Sq Ft")
plt.grid()
plt.legend()
plt.title('Graph 2  (Total square footage vs Congress Dist)')


# plotting a subplot for Average square footage per floor as a function of congressional district
plt.subplot(222)
plt.plot(df2['Congress Dist'],df2['Avg Sq Ft/Floor'],'C3',lw=0.5,marker = '*',label='Avg Sq Ft/Floor')
plt.xlabel("Congress Dist")
plt.ylabel("Avg Sq Ft/Floor")
plt.grid()
plt.legend()
plt.title('Graph 3  (Average square footage/floor vs Congress Dist)')


```

#### power 10^7), on the other hand, the Mean Square Footage per floor though high, but is numerically less when compared to the 

#### Total Footage
